## Determine a suitable merge method

```python
# Drop _merge column to enable new merge indicator test
df_ords_prods_co=df_ords_prods_co.drop(columns=['_merge'])
```

```python
# Test merge without overwriting
pd.merge(df_ords_prods_co,df_prods,on=['product_id'],indicator=True)
```

## Executing merge

```python
# Execute merge, add merge flag
df_ords_prods_merged=df_ords_prods_co.merge(df_prods,on=['product_id'],indicator=True)
```
